Dalam pemasaran produk skincare di segmen mahasiswa, salah satu tantangan yang sering dihadapi perusahaan adalah menentukan channel komunikasi yang paling efektif untuk membangun awareness sekaligus mendorong ketertarikan konsumen. Tanpa pemahaman yang tepat mengenai jalur ini, strategi pemasaran berisiko menjadi tidak efisien dan kurang tepat sasaran.

Untuk itu, dilakukan analisis Source of Awareness and Interest (SOA) terhadap mahasiswa Matematika UIN Syarif Hidayatullah Jakarta angkatan 2021–2025. Analisis ini bertujuan untuk mengidentifikasi pola perilaku mahasiswa dalam mengenal dan tertarik pada brand facial wash, sekaligus memberikan dasar yang relevan dapat dimanfaatkan oleh perusahaan sebagai bahan pertimbangan dalam merancang strategi pemasaran yang lebih efektif dan berbasis data.

# **METODE DAN TUJUAN**
* Metode : Analysis Source of Awareness & Interest
  <BR>Analisis SOA dilakukan dengan mengombinasikan dua perspektif utama, yaitu claimed awareness (sumber informasi yang secara langsung disebut oleh responden) dan derived importance (pengaruh masing-masing channel dalam mendorong ketertarikan dan keputusan penggunaan produk).
* Tujuan : Mengidentifikasi pola perilaku konsumen dalam mengenal dan tertarik pada suatu produk, serta sejauh mana channel tertentu memengaruhi ketertarikan mereka untuk mulai menggunakan produk tersebut.  
* Alasan : Analisis ini membantu memahami jalur komunikasi yang efektif bagi brand dalam membangun awareness dan interest di segmen konsumen tertentu.

```python
# List produk yang ingin dibreakdown
brand_soa = ["Scora", "Wardah", "Senka", "Cetaphil", "Glad2Glow"]

# List platform (dari df.columns kamu)
platform_soa = [
    "Teman/Keluarga", "Instagram", "TikTok", "X",
    "Iklan di TV/Radio", "Poster/Billboard/Brosur",
    "Shopee", "Tokopedia", "Rekomendasi dari toko kosmetik"
]

result = []

for idx, row in df2.iterrows():
    respondent = row["RespondentID"]
    brand_pakai = row["Brand"]   # brand yang digunakan

    for brand in brand_soa:
        row_dict = {
            "RespondentID": respondent,
            "Brand": brand,
            "Pakai": 1 if brand == brand_pakai else 0
        }

        # cek semua platform
        for p in platform_soa:
            sumber = str(row[p])  # ambil isi platform
            # isi 1 jika brand muncul di kolom platform tsb
            row_dict[p] = 1 if brand in sumber else 0

        result.append(row_dict)

df_breakdown = pd.DataFrame(result)
df_breakdown
```

#**Analisis SOA Secara Keseluruhan (Pengaruh Channel Terhadap Pemakaian Produk Facial Wash Secara Umum)**

##Claimed (Hitung Persentase “Dari Mana Kamu Mencari Informasi/Mengetahui")

```python
# Pastikan 0/1 dan tidak ada NaN
df_breakdown[platform_soa] = df_breakdown[platform_soa].fillna(0).astype(int)

# --- INTI PERHITUNGAN YANG BENAR ---

# 1) Naik ke level RESPONDEN:
#    per RespondentID, 1 kalau dia PERNAH pakai channel tsb untuk brand apapun
resp_level = (
    df_breakdown
    .groupby("RespondentID")[platform_soa]
    .max()
)

# 2) Total responden
total_resp = resp_level.shape[0]

# 3) Jumlah responden per channel
claimed_count = resp_level.sum()

# 4) Persentase responden per channel
claimed_pct = (claimed_count / total_resp * 100).round(1)

# 5) Susun tabel rapi
claimed_df = pd.DataFrame({
    "Channel": claimed_count.index,
    "Jumlah_Responden": claimed_count.values,
    "Claimed (%)": claimed_pct.values
}).sort_values("Claimed (%)", ascending=False)

print("Claimed SOA (jumlah & persentase responden per channel):")
print(claimed_df)

# (Opsional) Plot
plt.figure(figsize=(10, 5))
sns.barplot(data=claimed_df, x="Channel", y="Claimed (%)")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Persentase responden (%)")
plt.title("Claimed Source of Awareness (SOA)")
plt.tight_layout()
plt.show()
```

#### Dapat dilihat dari hasil claimed SOA Hasil analisis menunjukkan bahwa sumber informasi yang paling banyak disebut oleh responden adalah Teman/Keluarga dan TikTok, masing-masing dengan persentase 25,9%. Selain itu, Shopee (22,4%) dan Instgram (19%) juga berperan cukup besar sebagai sumber informasi dalam membangun awarness. Sementara itu, X menjadi sumber informasi yang paling jarang disebut, dengan persentase responden sebesar 1,7%.

##Derived Importance : Lakukan regresi logistik, Hitung odds ratio tiap faktor, Lakukan normalisasi

```python
# 1. Y = Pakai (0/1)
y = df_breakdown["Pakai"]

# 2. X = channel-channel SOA
X = df_breakdown[platform_soa]

# 3. Tambahkan intercept
X = sm.add_constant(X)

# 4. Fit model regresi logistik
model = sm.Logit(y, X).fit(disp=False)

# 5. Ambil koefisien (log-odds), buang intercept
coef = model.params[1:]

# 6. Hitung Odds Ratio
odds_ratio = np.exp(coef)

# 7. Normalisasi OR agar total = 100%
derived_importance = (odds_ratio / odds_ratio.sum() * 100).round(1)

# 8. Buat DataFrame rapi
derived_df = pd.DataFrame({
    "Channel": odds_ratio.index,
    "Odds Ratio": odds_ratio.values.round(3),
    "Derived Importance (%)": derived_importance.values
})

# Urutkan berdasarkan Derived Importance tertinggi
derived_df = derived_df.sort_values("Derived Importance (%)", ascending=False).reset_index(drop=True)

print("=== TABEL DERIVED IMPORTANCE SOA ===")
display(derived_df)
```

#### Berdasarkan analisis Derived Importance, Shopee menjadi kanal dengan pengaruh paling besar dalam mendorong penggunaan produk facial wash, dengan kontribusi sebesar 36,2%. Hal ini menunjukkan bahwa Shopee berperan sebagai faktor utama dalam memengaruhi keputusan konsumen.

#### Selain Shopee, Iklan di TV/Radio (18,0%), Poster/Billboard/Brosur (15,0%), serta Teman/Keluarga (14,7%) juga memiliki pengaruh yang cukup kuat, meskipun masih berada di bawah marketplace. Sementara itu, TikTok (6,8%), Rekomendasi dari toko kosmetik (5,2%), dan Instagram (3,5%) berperan sebagai kanal pendukung dengan pengaruh yang lebih terbatas.

#### Di sisi lain, Tokopedia (0,5%) dan X (Twitter) menunjukkan kontribusi yang sangat minimal. Secara keseluruhan, hasil ini menegaskan bahwa marketplace—khususnya Shopee—memiliki peran yang lebih dominan dibandingkan media sosial dalam mendorong penggunaan produk facial wash.


```python
# Pastikan nama kolom sama
claimed_df["Channel"] = claimed_df["Channel"].str.strip()
derived_df["Channel"] = derived_df["Channel"].str.strip()

# Gabungkan Claimed + Derived
soa_quadrant = claimed_df.merge(
    derived_df[["Channel", "Derived Importance (%)"]],
    on="Channel",
    how="left"
)

# === 3. Hitung nilai tengah (median) ===
claimed_median = soa_quadrant["Claimed (%)"].median()
derived_median = soa_quadrant["Derived Importance (%)"].median()

print("Nilai Tengah (Median) Claimed  :", claimed_median)
print("Nilai Tengah (Median) Derived :", derived_median)
```

Setelah median didapat → Tentukan Kuadran

```python
def get_quadrant(row):
    c = row["Claimed (%)"]
    d = row["Derived Importance (%)"]

    if d >= derived_median and c >= claimed_median:
        return "High Derived - High Claimed (Maintain / Most Invest)"
    elif d >= derived_median and c < claimed_median:
        return "High Derived - Low Claimed (Opportunity to Optimize)"
    elif d < derived_median and c >= claimed_median:
        return "Low Derived - High Claimed (Occasionally)"
    else:
        return "Low Derived - Low Claimed (Last Priority)"

soa_quadrant["Quadrant"] = soa_quadrant.apply(get_quadrant, axis=1)

soa_quadrant
```

#**Analisis SOA Berdasarkan Brand (Pengaruh Channel Terhadap Pemilihan BRAND Tertentu)**

```python
def claimed_by_brand_all(df_breakdown, brand_soa, platform_soa):
    dfb = df_breakdown.copy()
    dfb[platform_soa] = dfb[platform_soa].fillna(0).astype(int)

    # Filter brand saja (tanpa filter Pakai)
    sub = dfb[dfb["Brand"] == brand_soa]

    resp_level = (
        sub
        .groupby("RespondentID")[platform_soa]
        .max()
    )

    total_resp = resp_level.shape[0]
    claimed_count = resp_level.sum()
    claimed_pct = (claimed_count / total_resp * 100).round(1)

    claimed_df1 = pd.DataFrame({
        "Channel": claimed_count.index,
        "Jumlah_Responden": claimed_count.values,
        "Claimed (%)": claimed_pct.values
    }).sort_values("Claimed (%)", ascending=False)

    return claimed_df1
```

```python
# Fungsi hitung derived importance
def derived_importance(df_breakdown, brand_soa, soa_cols):
    df_breakdown = df_breakdown.copy()
    df_breakdown["target"] = (df_breakdown["Brand"] == brand_soa).astype(int)

    X = df_breakdown[soa_cols]
    y = df_breakdown["target"]

    model = LogisticRegression(max_iter=500)

    model.fit(X, y)

    odds = np.exp(model.coef_[0])
    importance_raw = pd.Series(odds, index=soa_cols)

    importance_norm = importance_raw / importance_raw.sum()
    return importance_raw, importance_norm
```
